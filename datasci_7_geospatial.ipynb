# **Dataset 1: Obesity Precentages**

```python
df_obesity = gpd.read_file('Obesity_Percentages.geojson')
df_obesity
```

```python
df_obesity.columns
```

```python
missing_values = df_obesity.isnull().sum()
missing_values
```

```python
df_obesity[['Pct_Obese', 'ZIP']].sort_values(by=['Pct_Obese'], ascending=False)
```

```python
df_obesity.plot("Pct_Obese", legend=True)
```

## **Analysis Insights**

**Dataset Description:** The dataset contains obesity percentages for Lake County, Illinois.

**Attributes:**

- Pct_Obese: The percent of people in the zip code who are considered obese, defined as having a BMI greater than or equal to 30.

- ObsOrOvrwt: he percent of people in the zip code who are considered overweight (defined as having a BMI greater than or equal to 25 but less than 30) or obese (defined as having a BMI greater than or equal to 30).

**Patterns or Anomalies**
- The lower third of the country has much higher obesity rates than the it's two-thirds top counterparts.

**Visulization Decisions**
- I did not choose a specific color or theme, I left the map to be at the default theme.

## **Analysis Insights**

**Dataset Description:** The dataset contains births rates across Lake County, Illinois by ZIP Code.

**Attributes:**

- LBW - Low birth weight is defined as a birth where the baby weighs less than 2,500 grams. This is a percent.

- Preterm - Preterm birth is defined as a birth that occur before 37 weeks of pregnancy. This is a percent.

- TeenBirth: Teen births are defined as women aged 15 to 19 years who give birth. This is a rate.

- Birth_Rate: Birth rate is defined as the number of live births per 1,000 populations.

- F1stTriCare: 1st Trimester of care refers to the doctor’s visits and care provided during the first 13 weeks of pregnancy. This is a percent.

**Patterns or Anomalies**
- The rate of Teen Birth was very high on the north eastern side of the county, whereas much of the southern part of the county had lower rates

**Visulization Decisions**

# Dataset 4: Cancer Rates

```python
df_cancer = gpd.read_file('Cancer_Rates.geojson')
df_cancer
```

```python
missing_values = df_cancer.isnull().sum()
missing_values
```

```python
df_cancer.plot("Lung_Bronc", 'Blues', legend=True)
```

## **Analysis Insights**

**Dataset Description:** The dataset contains cancer Rates for Lake County Illinois.

**Attributes:**

- Colorectal Cancer - Cancer that develops in the colon (the longest part of the large intestine) and/or the rectum (the last several inches of the large intestine). This is a rate per 100,000.

- = Lung Cancer – Cancer that forms in tissues of the lung, usually in the cells lining air passages. This is a rate per 100,000.

- Breast Cancer – Cancer that forms in tissues of the breast. This is a rate per 100,000.       

- Prostate Cancer – Cancer that forms in tissues of the prostate. This is a rate per 100,000.

- Urinary System Cancer – Cancer that forms in the organs of the body that produce and discharge urine. These include the kidneys, ureters, bladder, and urethra. This is a rate per 100,000.

- All Cancer – All cancers including, but not limited to: colorectal cancer, lung cancer, breast cancer, prostate cancer, and cancer of the urinary system. This is a rate per 100,000.


**Patterns or Anomalies**
- Lung cancer rates are more prevalent on the borders of the county as compared to the middle of the county.

**Visulization Decisions**
- I chose to use variations of blues for the map.
